```python
import pandas as pd
data = pd.read_csv('marieboucher.csv', header=None)
data.dropna()
```

```python
nodeMap = {}

def getNode(row):
    if row[0]=='source' or pd.isnull(row[1])  or pd.isnull(row[0])  or pd.isnull(row[2]):
        return;
    if row[0] not in nodeMap:
        nodeMap[row[0]] = 0
    if row[2] not in nodeMap:
        nodeMap[row[2]] = 0
    nodeMap[row[0]] = nodeMap[row[0]] + 1 
    nodeMap[row[2]] = nodeMap[row[2]] + 1

data.apply(axis=1,func=getNode)

nodes = []
for idx, val in enumerate(list(nodeMap.items())):
    print (val)
    nodes.append({\
                  'id':val[0],
                  'label':val[0],
                  'weight':val[1],
                  'type':'People'
    })
```

```python
links = [] 
def getLink(row):
    if row[0]=='source' or pd.isnull(row[1])  or pd.isnull(row[0])  or pd.isnull(row[2]) or (row[0] not in nodeMap) or (row[2] not in nodeMap):
        return;
    date = int(row[3].split('/')[2])
    links.append({\
                  'id':int(row.name),\
                  'source':row[0],\
                  'target':row[2],\
                  'type':row[1],\
                  'start':date,\
                  'end':date\
                 })
    
data.apply(axis=1, func=getLink)

network = { 'nodes':nodes, 'links':links}
network
```

```python
import json
with open('marieboucher.json', 'w') as outfile:
    json.dump(network, outfile)
```
